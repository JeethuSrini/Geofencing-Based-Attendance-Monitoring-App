```python
import os
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN
from imutils.video import WebcamVideoStream, FPS
import cv2

# Function to save cropped images
# Function to save cropped images
def save_cropped_images(frame, faces, output_dir, image_count):
    for i, face in enumerate(faces):
        x, y, width, height = face['box']
        cropped_face = frame[y:y+height, x:x+width]
        filename = f'face_{image_count * len(faces) + i}.jpg'
        cv2.imwrite(os.path.join(output_dir, filename), cropped_face)
        
# Function to perform live face detection with MTCNN using imutils
def live_face_detection_imutils(confidence_threshold=0.95, num_images=10):
    # Create an MTCNN face detection model
    detector = MTCNN()

    # Open a connection to the webcam (usually index 0, but you may need to adjust)
    vs = WebcamVideoStream(src=0).start()

    # Start the FPS counter
    fps = FPS().start()

    # Counter for captured images
    image_count = 0

    # Create output directory to save the cropped images
    output_dir = 'cropped_images'
    os.makedirs(output_dir, exist_ok=True)

    while True:
        # Capture video frame-by-frame using imutils
        frame = vs.read()

        # Check if the frame is empty
        if frame is None:
            break

        # Detect faces in the frame
        faces = detector.detect_faces(frame)

        # Draw bounding boxes and confidence scores
        for face in faces:
            x, y, width, height = face['box']
            confidence = face['confidence']

            if confidence >= confidence_threshold:
                cv2.rectangle(frame, (x, y), (x + width, y + height), (255, 0, 0), 2)
                cv2.putText(frame, f'Confidence: {confidence:.2f}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

        # Display the frame
        cv2.imshow('Live Face Detection', frame)

        # Save cropped images
        if len(faces) > 0 and image_count < num_images:
             save_cropped_images(frame, faces, output_dir, image_count)
             image_count += 1

        # Update the FPS counter
        fps.update()

        # Break the loop if 'q' is pressed or required number of images captured
        if cv2.waitKey(1) & 0xFF == ord('q') or image_count >= num_images:
            break

    # Stop the FPS counter
    fps.stop()

    # Release the webcam and close the OpenCV windows
    vs.stop()
    cv2.destroyAllWindows()

# Call the live_face_detection_imutils function
live_face_detection_imutils()
```
